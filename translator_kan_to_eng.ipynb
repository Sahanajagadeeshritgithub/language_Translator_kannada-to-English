```python
# Hyperparameters
embedding_dim = 64
lstm_units = 128

# Encoder
encoder_inputs = tf.keras.layers.Input(shape=(max_kn_length,))
encoder_embedding = tf.keras.layers.Embedding(kn_vocab_size, embedding_dim)(encoder_inputs)
encoder_lstm = tf.keras.layers.LSTM(lstm_units, return_state=True)
encoder_outputs, state_h, state_c = encoder_lstm(encoder_embedding)
encoder_states = [state_h, state_c]

# Decoder
decoder_inputs = tf.keras.layers.Input(shape=(max_en_length - 1,))
decoder_embedding = tf.keras.layers.Embedding(en_vocab_size, embedding_dim)(decoder_inputs)
decoder_lstm = tf.keras.layers.LSTM(lstm_units, return_sequences=True, return_state=True)
decoder_outputs, _, _ = decoder_lstm(decoder_embedding, initial_state=encoder_states)
decoder_dense = tf.keras.layers.Dense(en_vocab_size, activation='softmax')
decoder_outputs = decoder_dense(decoder_outputs)

# Define the model
model = tf.keras.Model([encoder_inputs, decoder_inputs], decoder_outputs)

# Compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the model
model.fit([encoder_input_data, decoder_input_data], decoder_target_data, epochs=500, verbose=1)
```

```python
# Define inference models
encoder_model = tf.keras.Model(encoder_inputs, encoder_states)

decoder_state_input_h = tf.keras.layers.Input(shape=(lstm_units,))
decoder_state_input_c = tf.keras.layers.Input(shape=(lstm_units,))
decoder_states_inputs = [decoder_state_input_h, decoder_state_input_c]

decoder_outputs, state_h, state_c = decoder_lstm(decoder_embedding, initial_state=decoder_states_inputs)
decoder_states = [state_h, state_c]
decoder_outputs = decoder_dense(decoder_outputs)

decoder_model = tf.keras.Model([decoder_inputs] + decoder_states_inputs, [decoder_outputs] + decoder_states)

# Function to translate Kannada to English
def translate_sentence(input_text):
    input_seq = kn_tokenizer.texts_to_sequences([input_text])
    input_seq = pad_sequences(input_seq, maxlen=max_kn_length, padding='post')

    states_value = encoder_model.predict(input_seq)

    target_seq = np.zeros((1, 1))
    target_seq[0, 0] = en_tokenizer.word_index['<start>']

    stop_condition = False
    translated_sentence = ""

    while not stop_condition:
        output_tokens, h, c = decoder_model.predict([target_seq] + states_value)
        sampled_token_index = np.argmax(output_tokens[0, -1, :])
        sampled_word = en_tokenizer.index_word.get(sampled_token_index, '')

        if sampled_word == "<end>" or len(translated_sentence) > max_en_length:
            stop_condition = True
        else:
            translated_sentence += " " + sampled_word

        target_seq[0, 0] = sampled_token_index
        states_value = [h, c]

    return translated_sentence.strip()

# Test the model
test_sentences = ["ನಮಸ್ಕಾರ", "ಆರಾಮಾ?", "ನಾನು ಚೆನ್ನಾಗಿದ್ದೇನೆ", "ಸಹಾಯ ಬೇಕಾ?"]

for sentence in test_sentences:
    print(f"Input: {sentence} -> Output: {translate_sentence(sentence)}")
```

```python
# Install required libraries
!pip install beautifulsoup4 requests pandas googlesearch-python langdetect

import requests
from bs4 import BeautifulSoup
from googlesearch import search
import pandas as pd
import time
from langdetect import detect

# Function to check if a word is in Kannada script
def is_kannada(text):
    try:
        return detect(text) == "kn"
    except:
        return False

# Function to extract Kannada-English translations from a webpage
def extract_translations(url):
    try:
        headers = {"User-Agent": "Mozilla/5.0"}  # Prevent blocking
        response = requests.get(url, headers=headers)
        response.raise_for_status()  # Check for request errors
        soup = BeautifulSoup(response.text, "html.parser")

        translations = []

        # Extract text from <tr> and <td> in case it's a table
        for row in soup.find_all("tr"):
            cells = row.find_all("td")
            if len(cells) >= 2:
                kannada_text = cells[0].get_text(strip=True)
                english_text = cells[1].get_text(strip=True)
                if is_kannada(kannada_text):  # Ensure it's actually Kannada
                    translations.append((kannada_text, english_text))

        # Extract text from lists or paragraphs with patterns
        for li in soup.find_all("li"):
            text = li.get_text(strip=True)
            if " - " in text:  # Look for translation pattern
                parts = text.split(" - ")
                if len(parts) == 2 and is_kannada(parts[0]):
                    translations.append((parts[0], parts[1]))

        for p in soup.find_all("p"):  # Paragraphs may contain translations
            text = p.get_text(strip=True)
            if " - " in text:
                parts = text.split(" - ")
                if len(parts) == 2 and is_kannada(parts[0]):
                    translations.append((parts[0], parts[1]))

        return translations

    except Exception as e:
        print(f"Error processing {url}: {e}")
        return []

# Perform Google search for Kannada-English translations
query = "Kannada to English common words and phrases"
num_results = 5  # Number of search results to scrape

all_translations = []
for url in search(query, num_results=num_results):
    print(f"Scraping: {url}")
    translations = extract_translations(url)
    all_translations.extend(translations)
    time.sleep(2)  # Avoid sending too many requests quickly

# Convert extracted data into a Pandas DataFrame
df = pd.DataFrame(all_translations, columns=["Kannada", "English"])

# Remove any empty or malformed entries
df = df.dropna()
df = df[df["Kannada"] != ""]

# Save extracted data to a CSV file
df.to_csv("kannada_english_translations.csv", index=False)
print("Data saved successfully!")

# Display extracted translations
df.head()
```
